```python
import csv
from apify_client import ApifyClient
from datetime import datetime, timedelta
import pandas as pd
```

```python
client = ApifyClient('apify_api_12ke9PXisGi6cJz812x8GHsloIFFYh0itsxx')

start_date = datetime(2024, 12, 28)
end_date = datetime(2025, 1, 1)
```

```python
# Open a CSV file for writing
with open("goog_remaining_twitter_data.csv", "w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["date", "tweet_id", "username", "text", "likes", "retweets", "verified"])  # Headers

    current_date = start_date
    while current_date < end_date:
        next_date = current_date + timedelta(days=1)
        
        # Fetch tweets from verified users first
        verified_input = {
            "startUrls": ["https://x.com/search?q=%24PG%20OR%20%22proctor%20%26%20gamble%20earnings%22%20OR%20%22proctor%20%26%20gamble%20market%20outlook%22%20OR%20%22proctor%20%26%20gamble%20market%20performance%22%20OR%20%22proctor%20%26%20gamble%20price%20target%22%20OR%20%22proctor%20%26%20gamble%20share%20price%22%20OR%20%22proctor%20%26%20gamble%20stock%22%20OR%20%22proctor%20%26%20gamble%20financials%22%20OR%20%22proctor%20%26%20gamble%20dividend%22%20lang%3Aen&src=typed_query"],
            "onlyVerifiedUsers": True,
            "sort": "Top",
            "maxItems": 15,
            "start": current_date.strftime("%Y-%m-%d"),
            "end": next_date.strftime("%Y-%m-%d"),
        }

        print(f"Fetching verified tweets for {current_date.strftime('%Y-%m-%d')}...")
        verified_run = client.actor("nfp1fpt5gUlBwPcor").call(run_input=verified_input)

        tweets = []
        
        for item in client.dataset(verified_run["defaultDatasetId"]).iterate_items():
            tweets.append([
                item.get("createdAt"),  # Date
                item.get("id"),  # Tweet ID
                item.get("author", {}).get("userName"),  # Username
                item.get("fullText"),  # Tweet text
                item.get("likeCount"),  # Likes
                item.get("retweetCount"),  # Retweets
                True  # Verified user
            ])

        # Write the final list of tweets (max 15) to the CSV
        for tweet in tweets:
            writer.writerow(tweet)

        current_date = next_date  # Move to the next day
```
