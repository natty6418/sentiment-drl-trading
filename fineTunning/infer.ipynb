```python
import kagglehub
import pandas as pd
from pathlib import Path

def download_dataset(output_csv="stock_tweets.csv"):
    """
    Download dataset from Kaggle and save it as CSV
    Args:
        output_csv (str): Name of output CSV file
    Returns:
        Path: Path to the saved CSV file
    """
    try:
        # Download latest version
        path = kagglehub.dataset_download("equinxx/stock-tweets-for-sentiment-analysis-and-prediction")
        
        # Find and read the downloaded CSV
        csv_file = next(Path(path).glob("*.csv"))
        df = pd.read_csv(csv_file)
        
        # Save to local CSV
        output_path = Path(output_csv)
        df.to_csv(output_path, index=False)
        print(f"Dataset saved successfully to: {output_path.absolute()}")
        return output_path
        
    except Exception as e:
        print(f"Error processing dataset: {e}")
        return None
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM
from peft import PeftModel
from transformers import BitsAndBytesConfig
import pandas as pd
from datetime import datetime
from pathlib import Path

# === CONFIGURATION ===
use_huggingface_hub = True
hf_repo = "Natty6418/llama2-financial-sentiment"
base_model_name = "meta-llama/Llama-2-7b-hf"

# === TOKENIZER ===
tokenizer = AutoTokenizer.from_pretrained(hf_repo if use_huggingface_hub else base_model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# === MODEL LOADING WITH MODIFIED CONFIG ===
bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True
)

base_model = AutoModelForCausalLM.from_pretrained(
    base_model_name,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True
)

model = PeftModel.from_pretrained(base_model, hf_repo)
model.eval()

```

```python
torch.cuda.empty_cache()
```

```python
def get_sentiment(text, model, tokenizer):
    """Generate sentiment for a single tweet."""
    prompt = f"What is the sentiment of this tweet? Answer only with one word: positive, negative, or neutral.\nTweet: {text}\nSentiment:"
    inputs = tokenizer(prompt, return_tensors="pt", padding=True)
    with torch.no_grad():
        inputs = {key: value.to(model.device) for key, value in inputs.items()}
        outputs = model.generate(
            **inputs,
            max_new_tokens=50,
            temperature=0.1,
            top_p=0.95
        )
    return tokenizer.decode(outputs[0], skip_special_tokens=True).strip()

def process_dataset(model, tokenizer):
    """Process stock tweet dataset with sentiment analysis, saving checkpoints."""
    try:
        # Paths
        dataset_path = "./remaining.csv"
        checkpoint_path = "./sentiment_analysis_checkpoint_.csv"

        # Load dataset
        if Path(checkpoint_path).exists():
            print(f"Found checkpoint at {checkpoint_path}, resuming...")
            df = pd.read_csv(checkpoint_path)
        else:
            print("No checkpoint found, loading original dataset...")
            df = pd.read_csv(dataset_path)
            df['Sentiment'] = pd.NA  # Initialize Sentiment column

        # Verify required columns
        required_columns = ['Date', 'Tweet', 'Stock Name', 'Company Name']
        if not all(col in df.columns for col in required_columns):
            raise ValueError(f"Dataset missing required columns: {required_columns}")

        print(f"Processing {len(df)} tweets...")

        batch_size = 10
        already_processed = df['Sentiment'].notna()

        for i in range(0, len(df), batch_size):
            batch = df.iloc[i:i + batch_size]

            # Only process tweets that are not already processed
            unprocessed_batch = batch[~already_processed.iloc[i:i + batch_size]]
            if not unprocessed_batch.empty:
                df.loc[unprocessed_batch.index, 'Sentiment'] = unprocessed_batch['Tweet'].apply(
                    lambda text: get_sentiment(text, model, tokenizer)
                )
                print(f"Processed {min(i + batch_size, len(df))}/{len(df)} tweets")

                # Save after every batch
                df.to_csv(checkpoint_path, index=False)
                print(f"Checkpoint saved to {checkpoint_path}")

        # Save final results
        final_file = f"sentiment_analysis_final_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
        df.to_csv(final_file, index=False)
        print(f"\n✅ Final results saved to {final_file}")

        # Show sentiment distribution
        print("\n📊 Sentiment Distribution:")
        print(df['Sentiment'].value_counts())

        # Optionally delete checkpoint after success
        Path(checkpoint_path).unlink(missing_ok=True)
        print("\n🗑️ Deleted checkpoint file after successful completion.")

        return df

    except Exception as e:
        print(f"❌ Error during processing: {e}")
        raise
```

```python
process_dataset(model, tokenizer).to_csv("sentiment_analysis_final_.csv", index=False)
```

```python
stock_tweets_df = pd.read_csv("sentiment_analysis_final.csv")
```

```python
sentiment_map = {
    'positive': 1,
    'neutral': 0,
    'negative': -1
}

valid_sentiments = ["positive", "negative", "neutral"]

stock_tweets_df['Sentiment_Value'] = stock_tweets_df['Sentiment'].str.extract(r'Sentiment:\s*(\w+)')
stock_tweets_df = stock_tweets_df[stock_tweets_df['Sentiment_Value'].isin(valid_sentiments)]
stock_tweets_df['SentimentScore'] = stock_tweets_df['Sentiment_Value'].map(sentiment_map)
stock_tweets_df['Date'] = pd.to_datetime(stock_tweets_df['Date']).dt.date

# Group by date and stock
daily_score = stock_tweets_df.groupby(['Date', 'Stock Name'])['SentimentScore'].mean().reset_index()
```

```python
stock_tweets_df.head()
```

```python
num_positive_sentiments = daily_score[daily_score['SentimentScore'] >0].shape[0]
num_negative_sentiments = daily_score[daily_score['SentimentScore'] <0].shape[0]
num_neutral_sentiments = daily_score[daily_score['SentimentScore'] == 0].shape[0]

print(f"Number of positive sentiments: {num_positive_sentiments}")
print(f"Number of negative sentiments: {num_negative_sentiments}")
print(f"Number of neutral sentiments: {num_neutral_sentiments}")
```

```python
all_dates = daily_score['Date'].unique()
all_stocks = daily_score['Stock Name'].unique()

# Step 2: Create a complete index (cartesian product of all dates × all stocks)
import itertools
full_index = pd.DataFrame(list(itertools.product(all_dates, all_stocks)), columns=['Date', 'Stock Name'])

# Step 3: Merge with your existing sentiment score data
df_filled = pd.merge(full_index, daily_score, on=['Date', 'Stock Name'], how='left')

# Step 4: Fill missing sentiment values with 0
df_filled['SentimentScore'] = df_filled['SentimentScore'].fillna(0)

df_filled['Date'] = pd.to_datetime(df_filled['Date'])
df_filled['Date'] = df_filled['Date'] - pd.Timedelta(days=1)

# Step 5: Save to CSV
df_filled.to_csv("daily_sentiment_score.csv", index=False)
```
